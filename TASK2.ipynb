# DATA FILTERING

```python
# Example: Filter rows where 'ChipRate' is greater than a threshold, say 100
filtered_df = df[df['ChipRate'] > 100]

print("\nFiltered DataFrame (ChipRate > 100):")
print(filtered_df.head())
```

```python
# Filter rows where 'ChipRate' is greater than 100 and 'BlowFlow' is less than 200
filtered_df_multi = df[(df['ChipRate'] > 100) & (df['BlowFlow'] < 200)]
print("\nFiltered DataFrame (ChipRate > 100 and BlowFlow < 200):")
print(filtered_df_multi.head())
```

```python
# Example: Filter rows where 'ChipRate' is in a specific list of values
chiprate_list = [95, 105, 110]
filtered_df_in_list = df[df['ChipRate'].isin(chiprate_list)]
print("\nFiltered DataFrame (ChipRate in [95, 105, 110]):")
print(filtered_df_in_list.head())

# Filter rows where 'BlowFlow' is not in a specific list of values
blowflow_list = [200, 210]
filtered_df_not_in_list = df[~df['BlowFlow'].isin(blowflow_list)]
print("\nFiltered DataFrame (BlowFlow not in [200, 210]):")
print(filtered_df_not_in_list.head())
```

```python
# Example: Filter rows by their index values
index_list = [0, 2, 4]
filtered_df_index = df.loc[index_list]
print("\nFiltered DataFrame (Index in [0, 2, 4]):")
print(filtered_df_index)
```

# HANDLE MISSING VALUES

```python
# Example: Fill missing values with the mean of the column
df_filled = df.fillna(df.mean(numeric_only=True))

print("\nDataFrame with missing values filled:")
print(df_filled.head())

# Alternatively, drop rows with missing values
df_dropped = df.dropna()

print("\nDataFrame with missing values dropped:")
print(df_dropped.head())
```

# SUMMARY STATISTICS

```python
# Mean of numerical columns
mean_values = df.mean(numeric_only=True)
print("\nMean Values:")
print(mean_values)
```
